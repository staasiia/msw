```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Parameters
alpha = 0.4   # rybí porodnost
beta = 0.02   # míra predace tučňáků
gamma = 0.3   # úmrtnost tučňáků
delta = 0.01  # míra rozmnožování tučňáků

# Počáteční populace
x0 = 500  # fish
y0 = 20   # penguin

# Čas
t = np.linspace(0, 200, 1000)

# Lotka-Volterrovy rovnice
def lotka_volterra(X, t, alpha, beta, gamma, delta):
    x, y = X
    dxdt = alpha * x - beta * x * y  
    dydt = delta * x * y - gamma * y
    return [dxdt, dydt]

# řešit ODE
sol = odeint(lotka_volterra, [x0, y0], t, args=(alpha, beta, gamma, delta))
x, y = sol.T

# graf
plt.figure(figsize=(10,6))
plt.plot(t, x, label='Fish', linewidth=2)
plt.plot(t, y, label='Penguins', linewidth=2)
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('Lotka–Volterra Predator–Prey Model')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Parameters
alpha = 0.4   # rybí porodnost
beta = 0.02   # míra predace tučňáků
beta_h = 0.05 # míra predace lidí
gamma = 0.3   # úmrtnost tučňáků
delta = 0.01  # míra rozmnožování tučňáků
delta_h = 0.005 # míra lidské reprodukce
gamma_h = 0.1 # úmrtnost lidí

# Počáteční populace
x0 = 500  # fish
y0 = 20   # penguin
z0 = 10 # human

# Čas
t = np.linspace(0, 200, 1000)

# Lotka-Volterrovy rovnice
def lotka_volterra(X, t, alpha, beta, beta_h, gamma, gamma_h, delta, delta_h):
    x, y, z = X
    dxdt = alpha * x - beta * x * y - beta_h * x * z 
    dydt = delta * x * y - gamma * y
    dzdt = delta_h * x * z - gamma_h * z
    return [dxdt, dydt, dzdt]

# řešit ODE
sol = odeint(lotka_volterra, [x0, y0, z0], t, args=(alpha, beta, beta_h, gamma, gamma_h, delta, delta_h))
x, y, z = sol.T

# graf
plt.figure(figsize=(10,6))
plt.plot(t, x, label='Fish', linewidth=2)
plt.plot(t, y, label='Penguins', linewidth=2)
plt.plot(t, z, label='Humans', linewidth=2)
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('Lotka–Volterra Predator–Prey Model')
plt.legend()
plt.grid(True)
plt.show()

#Interpretace výsledků:
#V klasickém modelu (graf 1) vidíme pravidelné cykly mezi populacemi ryb a tučňáků.
#Přidáním lidí jako dalšího predátora (graf 2) dochází k prudkému poklesu tučňáků a narušení přirozené rovnováhy.
#Lidský lov výrazně ovlivňuje dynamiku modelu.
```
