```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.integrate import odeint

N = 1000 #zjednodušená verze celkové populace
I0, R0 = 1, 0 #1 osoba je infikována, 0 je uzdraveno
S0 = N - I0 - R0 #počet lidí, kteří mohou onemocnět

R0_odhad = 4.75 #Pravé neštovice=4.75
trvanlivost_nemoci = 7 
gamma = 1 / trvanlivost_nemoci
beta = R0_odhad * gamma

t = np.linspace(0, 160, 160) #simulace začíná dnem 0 a končí dnem 160 a bude mít celkem 160 bodů

def deriv(y, t, N, beta, gamma): #konstantní incidence (formule)
    S, I, R = y
    dS = -beta * S * I / N
    dI = beta * S * I / N - gamma * I
    dR = gamma * I 
    return dS, dI, dR 

y0 = S0, I0, R0 #počáteční podmínky (999, 1, 0)
ret = odeint(deriv, y0, t, args=(N, beta, gamma)) #funkce řeší, jak se SIR v čase mění podle derivační funkce
S, I, R = ret.T #dostaneme ty linky

plt.figure(figsize=(10,6)) #graf
plt.plot(t, S, 'b', label='Susceptible')
plt.plot(t, I, 'r', label='Infected')
plt.plot(t, R, 'g', label='Recovered')
plt.title('SIR Model Simulation')
plt.xlabel('Time (days)')
plt.ylabel('Number of People')
plt.legend()
plt.grid(True)
plt.show()

#otázky
peak_day = t[np.argmax(I)]
epidemic_duration = np.where(I < 1)[0]
end_day = epidemic_duration[0] if epidemic_duration.size > 0 else "Neskončilo"
final_infected = R[-1]
final_uninfected = S[-1]

print(f"Kdy dojde k vrcholu epidemie: Den {peak_day:.0f}")
print(f"Jak dlouho epidemie potrvá: Den {end_day}")
print(f"Kolik jedinců onemocnělo: {final_infected:.0f}")
print(f"Kolik jedinců neonemocnělo: {final_uninfected:.0f}")
```
